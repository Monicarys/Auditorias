```python
#FUNCIONES GENERALES
#Funcion para exportar archivos

def exportar_excel(df, nombre):
  nombre_archivo=nombre+'.xlsm'
  return df.to_excel(nombre_archivo)


#Definimos la funcion que nos cuenta los SI's
def contar_si(X):
  return (X=='si').sum().sum()+(X=='SI').sum().sum()+(X=='Si').sum().sum()+(X=='sI').sum().sum()+(X=='Sí').sum().sum()
#Definimos la funcion que nos cuenta los NO's
def contar_no(X):
  return (X=='no').sum().sum()+(X=='NO').sum().sum()+(X=='No').sum().sum()+(X=='nO').sum().sum()
#Funcion para calcular el porcentaje

def contar_si_no(X):
  return contar_si(X)+contar_no(X)

def porcentaje_si(X):
  return round((contar_si(X)/contar_si_no(X))*100,2)

def porcentaje_no(X):
  return round((contar_no(X)/contar_si_no(X))*100,2)


#Definimos la funcion de leer y limpiar archivo
def leer_limpiar(archivo):
  df=pd.read_excel(archivo)
#Para quitar los Nan ocupamos la siguiente funcion
  df.fillna('', inplace=True)
#Intentando quitar \n, pero aun no sale
  return df
```

```python
#Funciones para obtener el nombre y tipo de entidad
def nombre_entidad(archivo):
  return (str(archivo)[:-5])[2:]

def tipo_entidad(archivo):
  inicial=str(archivo)[:2]
  if inicial=='D_':
    return 'DIF'
  if inicial=='I_':
    return'IMCUFIDE'
  if inicial=='IM':
    return 'IMJUVE'
  if inicial=='O_':
    return 'ODAS'
  if inicial=='M_':
    return 'MUNICIPIO'
  #print("Error en el archivo", archivo)
  #raise ValueError
```

```python

#Funcion para encontrar la dataframe filtrada en la que aparecen los no
#Depende del archivo y el modulo Y LA FUNCION crear_modulo

def filtro_no_modulo(archivo, modulo):
  M = crear_modulo(archivo, modulo)
  if modulo=='M1':
    mask_M=(M[['PDF Oct','XLS Oct','XML o TXT Oct', 'PDF Nov', 'XLS Nov','XML o TXT Nov','PDF Dic','XLS Dic','XML o TXT Dic']] ==('No'or 'no' or 'NO' or 'nO')).any(axis=1)
    return M[mask_M]
  if modulo=='M2':
    mask_M=(M[['PDF Oct','TXT Oct','Oct',' PDF Nov','TXT Nov','Nov','PDF Dic', 'TXT Dic','Dic']] ==('No'or 'no'or 'NO'or 'nO')).any(axis=1)
    return M[mask_M]
  if modulo=='M3':
    mask_M=(M[['6Octu','7Oct','8Oct',' 9Nov','10Nov','11Nov','PDF Dic','TXT Dic','14Dic']] ==('No'or 'no'or 'NO'or 'nO')).any(axis=1)
    return M[mask_M]
  if modulo=='M4':
    mask_M=(M[['PDF Oct', 'XLS Oct','TXT Oct','PDF Nov','XLS Nov','TXT Nov','PDF Dic', 'XLS Dic','TXT Dic']] ==('No'or 'no'or 'NO'or 'nO')).any(axis=1)
    return M[mask_M]

#Funcion que nos devuelve la tabla de los incumplimientos del Modulo X de un archivo, depende de las DOS FUNCIONES ANTERIORES
def archivos_incumplimiento_modulo_entidad(archivo,modulo):
  M = crear_modulo(archivo, modulo)
  filtered_M_no = filtro_no_modulo(archivo,modulo)
#Para esto creemos una lista de los nombres de las columnas
  columnas=filtered_M_no.columns.tolist()
#Creemos una lista de los nombres de las filas
  filas=filtered_M_no.index.tolist()
  my_columns=['Entidad', 'Tipo', 'Modulo', 'Concepto', 'Archivo al que corresponde']
#definimos nuestra primer fila como vacia 
  my_data = np.array([['','','','','']])
  for i in filas:
    for j in columnas:
      if filtered_M_no.loc[i,j]=='No':
        lo_de_medio= np.array([[nombre_entidad(archivo),
                           tipo_entidad(archivo),
                           modulo,
                           filtered_M_no.loc[i,'Concepto'],
                           j       
                           ]])
        my_data=np.append (my_data, lo_de_medio, axis = 0)

  my_dataframe = pd.DataFrame(data=my_data, columns=my_columns)
  my_dataframe= my_dataframe.drop([0],axis=0)
  return my_dataframe


#Funcion que nos devuelve la tabla de los incumplimientos del Modulo X de todos los archivos
def archivos_incumplimiento_modulo_entidades(archivos, modulo):
  #Definimos la lista de los archivos de incumplimiento de cada archivo
  AIM1=[archivos_incumplimiento_modulo_entidad(archivo,modulo) for archivo in archivos]
  #Concatenamos verticalmente en una sola dataframe
  vertical_stack = pd.concat(AIM1, axis=0)
  return vertical_stack



```

```python
def analisis_general(archivos):


  #Creamos nuestras columnas
  columnas_ct=['Entidad', 'Tipo', 'Total de no', 'Total de si', 'Pocentaje de incumplimiento Trimestral','Pocentaje de cumplimiento Trimestral']
  #definimos nuestra primer fila como vacia 
  my_data_ct = np.array([['','','','','','' ]])

  for archivo in archivos:
    df=leer_limpiar(archivo)
    #Creamos nuestros modulos
    M1=crear_M1(df)
    M2=crear_M2(df)
    M3=crear_M3(df)
    M4=crear_M4(df)

  
    #Definimos algunas variables que ocuparemos
    sis_trimestre=contar_si(M1)+contar_si(M2)+contar_si(M3)+contar_si(M4)
    nos_trimestre=contar_no(M1)+contar_no(M2)+contar_no(M3)+contar_no(M4)
    siyno_trimestre=sis_trimestre + nos_trimestre
    porcentaje_si_trimestre=round((sis_trimestre/(siyno_trimestre)*100),2)
    porcentaje_no_trimestre=round((nos_trimestre/(siyno_trimestre)*100),2)

    #Vamos agregando cada fila con la informacion de cada modulo
    my_data_ct=np.append (my_data_ct, np.array([[nombre_entidad(archivo),
                           tipo_entidad(archivo), nos_trimestre, sis_trimestre, porcentaje_no_trimestre, porcentaje_si_trimestre]]), axis = 0)


  #Hacemos nuestro dataframe con tan informacion
  df_ct = pd.DataFrame(data=my_data_ct, columns=columnas_ct)
  return df_ct
```

#**3.¿A qué corresponde el incumplimiento de las entidades fiscalizables?**

**3.1 Archivos de incumplimiento entidad-modulo**

Existen varias maneras de analizar a que archivos corresponden los incumplimientos. Comezaremos desde lo particular e iremos a lo general

```python
#Funcion para encontrar la dataframe filtrada en la que aparecen los no
#Depende del archivo y el modulo

def filtro_no_modulo(archivo, modulo):
  M = crear_modulo(archivo, modulo)
  if modulo=='M1':
    mask_M=(M[['PDF Oct','XLS Oct','XML o TXT Oct', 'PDF Nov', 'XLS Nov','XML o TXT Nov','PDF Dic','XLS Dic','XML o TXT Dic']] ==('No'or 'no' or 'NO' or 'nO')).any(axis=1)
    return M[mask_M]
  if modulo=='M2':
    mask_M=(M[['PDF Oct','TXT Oct','Oct',' PDF Nov','TXT Nov','Nov','PDF Dic', 'TXT Dic','Dic']] ==('No'or 'no'or 'NO'or 'nO')).any(axis=1)
    return M[mask_M]
  if modulo=='M3':
    mask_M=(M[['6Octu','7Oct','8Oct',' 9Nov','10Nov','11Nov','PDF Dic','TXT Dic','14Dic']] ==('No'or 'no'or 'NO'or 'nO')).any(axis=1)
    return M[mask_M]
  if modulo=='M4':
    mask_M=(M[['PDF Oct', 'XLS Oct','TXT Oct','PDF Nov','XLS Nov','TXT Nov','PDF Dic', 'XLS Dic','TXT Dic']] ==('No'or 'no'or 'NO'or 'nO')).any(axis=1)
    return M[mask_M]

def archivos_incumplimiento_modulo_entidad(archivo,modulo):
  M = crear_modulo(archivo, modulo)
  filtered_M_no = filtro_no_modulo(archivo,modulo)
#Para esto creemos una lista de los nombres de las columnas
  columnas=filtered_M_no.columns.tolist()
#Creemos una lista de los nombres de las filas
  filas=filtered_M_no.index.tolist()
  my_columns=['Entidad', 'Tipo', 'Modulo', 'Concepto', 'Archivo al que corresponde']
#definimos nuestra primer fila como vacia 
  my_data = np.array([['','','','','']])
  for i in filas:
    for j in columnas:
      if filtered_M_no.loc[i,j]=='No':
        lo_de_medio= np.array([[nombre_entidad(archivo),
                           tipo_entidad(archivo),
                           modulo,
                           filtered_M_no.loc[i,'Concepto'],
                           j       
                           ]])
        my_data=np.append (my_data, lo_de_medio, axis = 0)

  my_dataframe = pd.DataFrame(data=my_data, columns=my_columns)
  my_dataframe= my_dataframe.drop([0],axis=0)
  return my_dataframe
```

Con la siguiente funcion obtenemos los archivos de incumplimiento de una entidad fiscalizable y correspondientes a un modulo.
