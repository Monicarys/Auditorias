```python
import pandas as pd
import numpy as np
import os
#Muestra los archivos
archivos=[file for file in os.listdir() if ".xlsm" in file]
archivos
```

```python
#FUNCIONES GENERALES
#Funcion para exportar archivos

def exportar_excel(df, nombre):
  nombre_archivo=nombre+'.xlsm'
  return df.to_excel(nombre_archivo)


#Definimos la funcion que nos cuenta los SI's
def contar_si(X):
  return (X=='si').sum().sum()+(X=='SI').sum().sum()+(X=='Si').sum().sum()+(X=='sI').sum().sum()+(X=='Sí').sum().sum()
#Definimos la funcion que nos cuenta los NO's
def contar_no(X):
  return (X=='no').sum().sum()+(X=='NO').sum().sum()+(X=='No').sum().sum()+(X=='nO').sum().sum()
#Funcion para calcular el porcentaje

def contar_si_no(X):
  return contar_si(X)+contar_no(X)

def porcentaje_si(X):
  return round((contar_si(X)/contar_si_no(X))*100,2)

def porcentaje_no(X):
  return round((contar_no(X)/contar_si_no(X))*100,2)


#Definimos la funcion de leer y limpiar archivo
def leer_limpiar(archivo):
  df=pd.read_excel(archivo)
#Para quitar los Nan ocupamos la siguiente funcion
  df.fillna('', inplace=True)
#Intentando quitar \n, pero aun no sale
  return df
```

```python

#POSIBLEMENTE PARA BORRAR ---Funcion para crear modulos 
def crear_M1(df):
  M1=df.iloc[10:49, [1,2,3,6,7,8,9,10,11,12,13,14]] 
  M1.rename(columns={'Unnamed: 1':'Submodulo', 'Unnamed: 2':'N','Unnamed: 3':'Concepto', 'Unnamed: 6':'PDF Oct', 'Unnamed: 7':'XLS Oct','Unnamed: 8':'XML o TXT Oct', 'Unnamed: 9':'PDF Nov', 'Unnamed: 10':'XLS Nov','Unnamed: 11':'XML o TXT Nov', 'Unnamed: 12':'PDF Dic', 'Unnamed: 13':'XLS Dic','Unnamed: 14':'XML o TXT Dic'}, inplace=True)
  return M1

def crear_M2(df):
  M2=df.iloc[57:81, [1,2,3,6,7,8,9,10,11,12,13,14]]
  M2.rename(columns={'Unnamed: 1':'Submodulo', 'Unnamed: 2':'N','Unnamed: 3':'Concepto', 'Unnamed: 6':'PDF Oct', 'Unnamed: 7':'TXT Oct','Unnamed: 8':'Oct', 'Unnamed: 9':' PDF Nov', 'Unnamed: 10':'TXT Nov','Unnamed: 11':'Nov', 'Unnamed: 12':'PDF Dic', 'Unnamed: 13':'TXT Dic','Unnamed: 14':'Dic'}, inplace=True)
  return M2

def crear_M3(df):
  M3=df.iloc[88:98, [1,2,3,6,7,8,9,10,11,12,13,14]]
  M3.rename(columns={'Unnamed: 1':'Submodulo', 'Unnamed: 2':'N','Unnamed: 3':'Concepto', 'Unnamed: 6':'6Octu', 'Unnamed: 7':'7Oct','Unnamed: 8':'8Oct', 'Unnamed: 9':' 9Nov', 'Unnamed: 10':'10Nov','Unnamed: 11':'11Nov', 'Unnamed: 12':'PDF Dic', 'Unnamed: 13':'TXT Dic','Unnamed: 14':'14Dic'}, inplace=True)
  return M3

def crear_M4(df):
  M4=df.iloc[105:128, [1,2,3,6,7,8,9,10,11,12,13,14]]
  M4.rename(columns={'Unnamed: 1':'Submodulo', 'Unnamed: 2':'N','Unnamed: 3':'Concepto', 'Unnamed: 6':'PDF Oct', 'Unnamed: 7':'XLS Oct','Unnamed: 8':'TXT Oct', 'Unnamed: 9':'PDF Nov', 'Unnamed: 10':'XLS Nov','Unnamed: 11':'TXT Nov', 'Unnamed: 12':'PDF Dic', 'Unnamed: 13':'XLS Dic','Unnamed: 14':'TXT Dic'}, inplace=True)
  return M4

```

```python

#Funciones para obtener el nombre y tipo de entidad
def nombre_entidad(archivo):
  return (str(archivo)[:-5])[2:]

def tipo_entidad(archivo):
  inicial=str(archivo)[:2]
  if inicial=='D_':
    return 'DIF'
  if inicial=='I_':
    return'IMCUFIDE'
  if inicial=='IM':
    return 'IMJUVE'
  if inicial=='O_':
    return 'ODAS'
  if inicial=='M_':
    return 'MUNICIPIO'
  print("Error en el archivo", archivo)
  raise ValueError
```

```python
#Funcion para crear modulos 
def crear_modulo(archivo, modulo):
  df=leer_limpiar(archivo)
  if modulo=='M1':
    M1=df.iloc[10:49, [1,2,3,6,7,8,9,10,11,12,13,14]] 
    M1.rename(columns={'Unnamed: 1':'Submodulo', 'Unnamed: 2':'N','Unnamed: 3':'Concepto', 'Unnamed: 6':'PDF Oct', 'Unnamed: 7':'XLS Oct','Unnamed: 8':'XML o TXT Oct', 'Unnamed: 9':'PDF Nov', 'Unnamed: 10':'XLS Nov','Unnamed: 11':'XML o TXT Nov', 'Unnamed: 12':'PDF Dic', 'Unnamed: 13':'XLS Dic','Unnamed: 14':'XML o TXT Dic'}, inplace=True)
    return M1

  if modulo=='M2':
    M2=df.iloc[57:81, [1,2,3,6,7,8,9,10,11,12,13,14]]
    M2.rename(columns={'Unnamed: 1':'Submodulo', 'Unnamed: 2':'N','Unnamed: 3':'Concepto', 'Unnamed: 6':'PDF Oct', 'Unnamed: 7':'TXT Oct','Unnamed: 8':'Oct', 'Unnamed: 9':' PDF Nov', 'Unnamed: 10':'TXT Nov','Unnamed: 11':'Nov', 'Unnamed: 12':'PDF Dic', 'Unnamed: 13':'TXT Dic','Unnamed: 14':'Dic'}, inplace=True)
    return M2

  if modulo=='M3':
    M3=df.iloc[88:98, [1,2,3,6,7,8,9,10,11,12,13,14]]
    M3.rename(columns={'Unnamed: 1':'Submodulo', 'Unnamed: 2':'N','Unnamed: 3':'Concepto', 'Unnamed: 6':'6Octu', 'Unnamed: 7':'7Oct','Unnamed: 8':'8Oct', 'Unnamed: 9':' 9Nov', 'Unnamed: 10':'10Nov','Unnamed: 11':'11Nov', 'Unnamed: 12':'PDF Dic', 'Unnamed: 13':'TXT Dic','Unnamed: 14':'14Dic'}, inplace=True)
    return M3

  if modulo=='M4':
    M4=df.iloc[105:128, [1,2,3,6,7,8,9,10,11,12,13,14]]
    M4.rename(columns={'Unnamed: 1':'Submodulo', 'Unnamed: 2':'N','Unnamed: 3':'Concepto', 'Unnamed: 6':'PDF Oct', 'Unnamed: 7':'XLS Oct','Unnamed: 8':'TXT Oct', 'Unnamed: 9':'PDF Nov', 'Unnamed: 10':'XLS Nov','Unnamed: 11':'TXT Nov', 'Unnamed: 12':'PDF Dic', 'Unnamed: 13':'XLS Dic','Unnamed: 14':'TXT Dic'}, inplace=True)
    return M4

#Funcion para encontrar la dataframe filtrada en la que aparecen los no
#Depende del archivo y el modulo Y LA FUNCION crear_modulo

def filtro_no_modulo(archivo, modulo):
  M = crear_modulo(archivo, modulo)
  if modulo=='M1':
    mask_M=(M[['PDF Oct','XLS Oct','XML o TXT Oct', 'PDF Nov', 'XLS Nov','XML o TXT Nov','PDF Dic','XLS Dic','XML o TXT Dic']] ==('No'or 'no' or 'NO' or 'nO')).any(axis=1)
    return M[mask_M]
  if modulo=='M2':
    mask_M=(M[['PDF Oct','TXT Oct','Oct',' PDF Nov','TXT Nov','Nov','PDF Dic', 'TXT Dic','Dic']] ==('No'or 'no'or 'NO'or 'nO')).any(axis=1)
    return M[mask_M]
  if modulo=='M3':
    mask_M=(M[['6Octu','7Oct','8Oct',' 9Nov','10Nov','11Nov','PDF Dic','TXT Dic','14Dic']] ==('No'or 'no'or 'NO'or 'nO')).any(axis=1)
    return M[mask_M]
  if modulo=='M4':
    mask_M=(M[['PDF Oct', 'XLS Oct','TXT Oct','PDF Nov','XLS Nov','TXT Nov','PDF Dic', 'XLS Dic','TXT Dic']] ==('No'or 'no'or 'NO'or 'nO')).any(axis=1)
    return M[mask_M]

#Funcion que nos devuelve la tabla de los incumplimientos del Modulo 1 de un archivo, depende de las DOS FUNCIONES ANTERIORES
def archivos_incumplimiento_modulo_entidad(archivo,modulo):
  M = crear_modulo(archivo, modulo)
  filtered_M_no = filtro_no_modulo(archivo,modulo)
#Para esto creemos una lista de los nombres de las columnas
  columnas=filtered_M_no.columns.tolist()
#Creemos una lista de los nombres de las filas
  filas=filtered_M_no.index.tolist()
  my_columns=['Entidad', 'Tipo', 'Modulo', 'Concepto', 'Archivo al que corresponde']
#definimos nuestra primer fila como vacia 
  my_data = np.array([['','','','','']])
  for i in filas:
    for j in columnas:
      if filtered_M_no.loc[i,j]=='No':
        lo_de_medio= np.array([[nombre_entidad(archivo),
                           tipo_entidad(archivo),
                           modulo,
                           filtered_M_no.loc[i,'Concepto'],
                           j       
                           ]])
        my_data=np.append (my_data, lo_de_medio, axis = 0)

  my_dataframe = pd.DataFrame(data=my_data, columns=my_columns)
  my_dataframe= my_dataframe.drop([0],axis=0)
  return my_dataframe


#Funcion que nos devuelve la tabla de los incumplimientos del Modulo 1 de todos los archivos
def archivos_incumplimiento_modulo_entidades(archivos, modulo):
  #Definimos la lista de los archivos de incumplimiento de cada archivo
  AIM1=[archivos_incumplimiento_modulo_entidad(archivo,modulo) for archivo in archivos]
  #Concatenamos verticalmente en una sola dataframe
  vertical_stack = pd.concat(AIM1, axis=0)
  return vertical_stack



```

```python
#Funcion para obtener el analisis mas general del recuento de sis y nos
def analisis_general(archivos):


  #Creamos nuestras columnas
  columnas_ct=['Entidad', 'Tipo', 'Total de no', 'Total de si', 'Pocentaje de incumplimiento Trimestral','Pocentaje de cumplimiento Trimestral']
  #definimos nuestra primer fila como vacia 
  my_data_ct = np.array([['','','','','','' ]])

  for archivo in archivos:
    df=leer_limpiar(archivo)
    #Creamos nuestros modulos
    M1=crear_M1(df)
    M2=crear_M2(df)
    M3=crear_M3(df)
    M4=crear_M4(df)

  
    #Definimos algunas variables que ocuparemos
    sis_trimestre=contar_si(M1)+contar_si(M2)+contar_si(M3)+contar_si(M4)
    nos_trimestre=contar_no(M1)+contar_no(M2)+contar_no(M3)+contar_no(M4)
    siyno_trimestre=sis_trimestre + nos_trimestre
    porcentaje_si_trimestre=round((sis_trimestre/(siyno_trimestre)*100),2)
    porcentaje_no_trimestre=round((nos_trimestre/(siyno_trimestre)*100),2)

    #Vamos agregando cada fila con la informacion de cada modulo
    my_data_ct=np.append (my_data_ct, np.array([[nombre_entidad(archivo),
                           tipo_entidad(archivo), nos_trimestre, sis_trimestre, porcentaje_no_trimestre, porcentaje_si_trimestre]]), axis = 0)


  #Hacemos nuestro dataframe con tan informacion
  df_ct = pd.DataFrame(data=my_data_ct, columns=columnas_ct)
  return df_ct
```

#**2. Analisis por modulo de las entidades fiscalizables** 






**2.1 Cumplimiento e incumplimiento un modulo**



El cumplimiento de un modulo en particular de las entidades fiscalizables se puede calcular utilizando la siguiente funcion

```python
#Hacemos nuestro dataframe con tal informacion

def analisis_modulo(archivos,modulo):

  #Creamos nuestras columnas
  columnas_cM=['Entidad', 'Tipo', 'Módulo' , 'Total de no','Total de si', 'Pocentaje de incumplimiento', 'Pocentaje de cumplimiento']
  #definimos nuestra primer fila como vacia 
  my_data_cM = np.array([['','','','','','','' ]])

  for archivo in archivos:
    M=crear_modulo(archivo, modulo)

  #Vamos agregando cada fila con la informacion de cada modulo
    my_data_cM=np.append (my_data_cM, np.array([[nombre_entidad(archivo),
                         tipo_entidad(archivo), modulo,contar_no(M), contar_si(M), porcentaje_no(M),porcentaje_si(M) ]]), axis = 0)
  
  df_cM = pd.DataFrame(data=my_data_cM, columns=columnas_cM)
  df_cM= df_cM.drop([0],axis=0)
  
  return df_cM
```

```python
analisis_modulo(archivos, 'M2')
```

**2.1 Cumplimiento e incumplimiento un modulo**



El cumplimiento e incumplimiento de varios modulos de las entidades fiscalizables se puede calcular utilizando la siguiente funcion

```python
#Funcion que nos devuelve el analisis de todos los modulos de las entidades fiscalizables
def analisis_modulos(archivos, modulos):
  #Definimos la lista de los archivos de incumplimiento de cada archivo
  AM=[analisis_modulo(archivos,modulo) for modulo in modulos]
  #Concatenamos verticalmente en una sola dataframe
  vertical_stack = pd.concat(AM, axis=0)
  return vertical_stack

```

```python
analisis_modulos(archivos,['M1','M2','M3','M4'])
```

```python
#Exportar a un excel 
exportar_excel(analisis_modulos(archivos,modulos), 'analisis_modulos_entidades')
```

#**3.¿A qué corresponde el incumplimiento de las entidades fiscalizables?**

**3.1 Archivos de incumplimiento entidad-modulo**

Existen varias maneras de analizar a que archivos corresponden los incumplimientos. Comezaremos desde lo particular e iremos a lo general

```python
#Funcion para encontrar la dataframe filtrada en la que aparecen los no
#Depende del archivo y el modulo

def filtro_no_modulo(archivo, modulo):
  M = crear_modulo(archivo, modulo)
  if modulo=='M1':
    mask_M=(M[['PDF Oct','XLS Oct','XML o TXT Oct', 'PDF Nov', 'XLS Nov','XML o TXT Nov','PDF Dic','XLS Dic','XML o TXT Dic']] ==('No'or 'no' or 'NO' or 'nO')).any(axis=1)
    return M[mask_M]
  if modulo=='M2':
    mask_M=(M[['PDF Oct','TXT Oct','Oct',' PDF Nov','TXT Nov','Nov','PDF Dic', 'TXT Dic','Dic']] ==('No'or 'no'or 'NO'or 'nO')).any(axis=1)
    return M[mask_M]
  if modulo=='M3':
    mask_M=(M[['6Octu','7Oct','8Oct',' 9Nov','10Nov','11Nov','PDF Dic','TXT Dic','14Dic']] ==('No'or 'no'or 'NO'or 'nO')).any(axis=1)
    return M[mask_M]
  if modulo=='M4':
    mask_M=(M[['PDF Oct', 'XLS Oct','TXT Oct','PDF Nov','XLS Nov','TXT Nov','PDF Dic', 'XLS Dic','TXT Dic']] ==('No'or 'no'or 'NO'or 'nO')).any(axis=1)
    return M[mask_M]

def archivos_incumplimiento_modulo_entidad(archivo,modulo):
  M = crear_modulo(archivo, modulo)
  filtered_M_no = filtro_no_modulo(archivo,modulo)
#Para esto creemos una lista de los nombres de las columnas
  columnas=filtered_M_no.columns.tolist()
#Creemos una lista de los nombres de las filas
  filas=filtered_M_no.index.tolist()
  my_columns=['Entidad', 'Tipo', 'Modulo', 'Concepto', 'Archivo al que corresponde']
#definimos nuestra primer fila como vacia 
  my_data = np.array([['','','','','']])
  for i in filas:
    for j in columnas:
      if filtered_M_no.loc[i,j]=='No':
        lo_de_medio= np.array([[nombre_entidad(archivo),
                           tipo_entidad(archivo),
                           modulo,
                           filtered_M_no.loc[i,'Concepto'],
                           j       
                           ]])
        my_data=np.append (my_data, lo_de_medio, axis = 0)

  my_dataframe = pd.DataFrame(data=my_data, columns=my_columns)
  my_dataframe= my_dataframe.drop([0],axis=0)
  return my_dataframe
```

Con la siguiente funcion obtenemos los archivos de incumplimiento de una entidad fiscalizable y correspondientes a un modulo.

```python
archivo='M_Ocoyoacac.xlsm'
archivos_incumplimiento_modulo_entidad(archivo,'M1')
```

**3.2 Archivos de incumplimiento entidades-modulo**

Con la siguiente funcion podemos analizar los archivos de incumplimiento por modulo de varias entidades fiscalizables

```python
#Funcion que nos devuelve la tabla de los incumplimientos del Modulo 1 de todos los archivos
def archivos_incumplimiento_modulo_entidades(archivos, modulo):
  #Definimos la lista de los archivos de incumplimiento de cada archivo
  AIM1=[archivos_incumplimiento_modulo_entidad(archivo,modulo) for archivo in archivos]
  #Concatenamos verticalmente en una sola dataframe
  vertical_stack = pd.concat(AIM1, axis=0)
  return vertical_stack
```
